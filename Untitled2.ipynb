# **Daily Transaction Analysis**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Daily Household Transactions.csv")
df
```

```python
df.isnull().sum() #get the null values
```

```python
df["Mode"].value_counts() #get the values count of the mode column
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(data = df, x = "Mode", order =
              df["Mode"].value_counts().iloc[:3].index)
plt.show() #
```

```python
df["Category"].value_counts() #get the values count of the category column
```

```python
sns.countplot(data = df, x = "Category", order =
              df["Category"].value_counts().iloc[:5].index)
plt.show()
#The categories with the tallest bars are the most frequent transaction categories in your dataset.
```

```python
df["Subcategory"].value_counts() #get the values count of the subcategory column
```

```python
 plt.figure(figsize=(12,8))
 sns.countplot(data=df,x="Subcategory",order=
 df["Subcategory"].value_counts().iloc[:10].index)
 plt.xticks(rotation=90)
 plt.show()
 # it displays the frequency of the top 10 most common subcategories found in the daily household transaction data./
```
